# Handling the missing data

```python
import numpy as np
import pandas as pd
```

```python
iris_df = pd.read_csv('/content/iris.csv')

data_frame = iris_df.copy()
data_frame.head()
```

```python
data_frame.describe()
```

```python
data_frame.iloc[2:3,2:4] = np.nan
data_frame.head()
```

```python
data_frame.describe()
# petal.length and petal.width count is reduced to 149 due to NaN values
```

```python
# dropna to drop row with NaN value
data_frame.dropna(inplace = True)
data_frame.head()
```

```python
data_frame.reset_index(drop=True, inplace = True)
data_frame.head()
```
