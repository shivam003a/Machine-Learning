# MatplotLib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
x1 = [1,2,3,4,5]
y = [1,2,3,4,5]
y1 = [5,4,3,2,1]

plt.scatter(x,y)
plt.scatter(x1,y1)
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
x = [1,2,3,4,5]
x1 = [1,2,3,4,5]
y = [1,2,3,4,5]
y1 = [5,4,3,2,1]

plt.plot(x,y,'r-.')
plt.plot(x1,y1,'y--')
plt.show()
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = 2*(x**4) + 100

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y = 2*(x**4) + 100

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4

plt.plot(x,y, color= 'red', marker='+', linewidth=.75)
plt.show()
```

```python
x = np.arange(0,10,0.25)
y = x**4
y2 = x**3

plt.plot(x,y, color= 'red', label='x^4')
plt.plot(x,y2, color= 'blue', label='x^3')
plt.xlabel('x', fontsize= 15)
plt.ylabel('y', fontsize = 15)
plt.title('X vs Y Graph')
plt.axis([0,20,0,10000])
plt.text(15,6000,'Random Text', fontsize = 10)
plt.grid()
plt.legend()
plt.show()
```
